Importing libraries.

```python
#Import Libraries.
import pandas as pd
```

Uploading the data.

```python
#Loading Karamoja District data.
District_data=pd.read_csv('/content/Uganda_Karamoja_District_Crop_Yield_Population.csv')

#Loading Karamoja Subcounty data.
Subcounty_data=pd.read_csv('/content/Uganda_Karamoja_Subcounty_Crop_Yield_Population.csv')
```

Preview of the Data.

```python
#Preview of the First five rows to confirm data is present.
District_data.head()
```

```python
#Preview of the First five rows to confirm data is present.
Subcounty_data.head()
```

Checking the range of the data.

```python
#Checking to see the values in the District dataset.
District_data.shape
District_data.info()
```

```python
#Checking to see the values in the Subcounty dataset.
Subcounty_data.shape
Subcounty_data.info()
```

Inspecting the data for any errors.

```python
#Checking for Null values.
District_data.isnull().sum()
```

```python
#Checking for null values in Sub county data.
Subcounty_data.isnull().sum()
```

```python
#Checking for duplicates in District data.
District_data.duplicated().sum()
```

```python
#Checking for duplicates in Subcounty data.
Subcounty_data.duplicated().sum()
```

Merging the data.

```python
#Merging the Subcounty data with the District data using 'Name' and 'District Name' columns
Karamoja_data=pd.merge(Subcounty_data,District_data,left_on='SUBCOUNTY_NAME',right_on='NAME',suffixes=('_subcounty', '_district'))
Karamoja_data.head()
```

Saving the data.

```python
#Saving the merged data.
Karamoja_data.to_csv('Karamoja_data.csv')
```
